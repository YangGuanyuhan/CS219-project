```python
import matplotlib.pyplot as plt
import numpy as np

# Data
vector_lengths = [10, 100, 1000, 10000]
c_times = [16095, 110264, 1289870, 10657444]
java_times = [165265, 325788, 1467086, 14370772]
ratios = [10.26, 2.95, 1.13, 1.34]

# Plot
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(vector_lengths, c_times, 'bo-', label='C Execution Time (us)')
ax1.plot(vector_lengths, java_times, 'ro-', label='Java Execution Time (us)')
ax2.plot(vector_lengths, ratios, 'go-', label='Java/C Time Ratio')

ax1.set_xlabel('Vector Length')
ax1.set_ylabel('Execution Time (us)')
ax2.set_ylabel('Java/C Time Ratio')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax2.set_yscale('log')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.title('C vs Java Execution Time Comparison')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据提取
vector_lengths = [10, 100, 1000, 10000]  # 向量长度
c_times = [30301, 228231, 2285849, 22894146]  # C 总执行时间（us）
java_times = [238953, 732595, 4325484, 33641110]  # Java 总执行时间（us）
time_ratios = [7.88, 3.20, 1.89, 1.46]  # Java/C 时间比率

fig, ax1 = plt.subplots(figsize=(8, 5))

# 绘制 C 和 Java 执行时间
ax1.plot(vector_lengths, c_times, 'bo-', label='C Execution Time (us)')
ax1.plot(vector_lengths, java_times, 'ro-', label='Java Execution Time (us)')
ax1.set_xscale('log')  # 对数尺度
ax1.set_yscale('log')
ax1.set_xlabel("Vector Length")
ax1.set_ylabel("Execution Time (us)")
ax1.legend()
ax1.grid()

# 共享 x 轴，添加 Java/C 时间比率的次坐标轴
ax2 = ax1.twinx()
ax2.plot(vector_lengths, time_ratios, 'g*-', label='Java/C Time Ratio')
ax2.set_ylabel("Time Ratio (Java / C)")
ax2.legend(loc='upper right')

plt.title("Performance Comparison float data: C vs Java Execution Time & Ratio")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
vector_lengths = [10, 100, 1000, 10000]
c_total_times = [29721, 228231, 2285849, 22894146]
java_total_times = [195362, 732595, 4325484, 33641110]
time_ratios = [6.57, 3.20, 1.89, 1.46]

# Plotting the data
fig, ax1 = plt.subplots()

ax1.set_xlabel('Vector Length')
ax1.set_ylabel('Execution Time (us)', color='tab:blue')
ax1.plot(vector_lengths, c_total_times, label='C Total Time', color='tab:blue', marker='o')
ax1.plot(vector_lengths, java_total_times, label='Java Total Time', color='tab:orange', marker='o')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for time ratio
ax2 = ax1.twinx()
ax2.set_ylabel('Java/C Time Ratio', color='tab:green')
ax2.plot(vector_lengths, time_ratios, label='Java/C Time Ratio', color='tab:green', marker='o')
ax2.tick_params(axis='y', labelcolor='tab:green')

# Title and layout
plt.title('Performance Comparison of C and Java (Data Type: float)')
fig.tight_layout()

# Show plot
plt.legend(loc='upper left')
plt.show()
```

```python
#optimal c vs original c
import matplotlib.pyplot as plt

# Data for the plot
test_cases = ['Test 1 (10)', 'Test 2 (100)', 'Test 3 (1000)', 'Test 4 (10000)']
original_times = [16105, 110398, 1161239, 11641687]
optimized_times = [2543, 48029, 115203, 43721]
performance_improvement = [6.33, 2.29, 10.07, 266.27]

# Creating subplots
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting the original and optimized times on the same graph
ax1.bar(test_cases, original_times, color='gray', alpha=0.5, label='Original Total Time (us)')
ax1.bar(test_cases, optimized_times, color='blue', alpha=0.5, label='Optimized Total Time (us)')

# Adding labels and title
ax1.set_xlabel('Test Cases')
ax1.set_ylabel('Execution Time (us)')
ax1.set_title('Execution Time Comparison (Original vs Optimized)')
ax1.legend()

# Create a second y-axis for performance improvement
ax2 = ax1.twinx()
ax2.plot(test_cases, performance_improvement, color='red', marker='o', label='Performance Improvement (x)', linestyle='--')
ax2.set_ylabel('Performance Improvement (x)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Display the plot
plt.tight_layout()
plt.show()
```
